```python
import pandas as pd
import matplotlib as plt

# Epsilon Vergleich
# X ist random, O ist Q-Learning

result_1 = pd.read_csv("results_of_E=0.1.csv", sep=";")
result_2 = pd.read_csv("results_of_E=0.2.csv", sep=";")
result_3 = pd.read_csv("results_of_E=0.3.csv", sep=";")
result_4 = pd.read_csv("results_of_E=0.4.csv", sep=";")
result_5 = pd.read_csv("results_of_E=0.5.csv", sep=";")
result_6 = pd.read_csv("results_of_E=0.6.csv", sep=";")
result_7 = pd.read_csv("results_of_E=0.7.csv", sep=";")
result_8 = pd.read_csv("results_of_E=0.8.csv", sep=";")
result_9 = pd.read_csv("results_of_E=0.9.csv", sep=";")
```

```python

# Füge die Ergebnisse zusammen spaltenweise zusammen
result = pd.concat([result_1, result_2, result_3, result_4, result_5, result_6, result_7, result_8, result_9], axis=1)

# Umbenennen der Spalten
result.columns = ["E=0.1", "E=0.2", "E=0.3", "E=0.4", "E=0.5", "E=0.6", "E=0.7", "E=0.8", "E=0.9"]

# Zähle die Anzahl der Gewinner für jede Einstellung von E
count_0_1 = result['E=0.1'].value_counts()
count_0_2 = result['E=0.2'].value_counts()
count_0_3 = result['E=0.3'].value_counts()
count_0_4 = result['E=0.4'].value_counts()
count_0_5 = result['E=0.5'].value_counts()
count_0_6 = result['E=0.6'].value_counts()
count_0_7 = result['E=0.7'].value_counts()
count_0_8 = result['E=0.8'].value_counts()
count_0_9 = result['E=0.9'].value_counts()

print(count_0_1)
print(count_0_9)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Melt, um das DataFrame für Seaborn vorzubereiten
df_melted = pd.melt(result, var_name='Einstellung von E', value_name='Gewinner')

# Plotte die Ergebnisse mit Seaborn
plt.figure(figsize=(8, 6))
sns.countplot(x='Gewinner', hue='Einstellung von E', data=df_melted, palette='husl')
plt.title('Ergebnisse für E=0.1 und E=0.9')
plt.ylabel('Anzahl')
plt.show()
```
